

# Standardization



***

<p>

<ul>
    
<li> When features of the input Dataset have large differences between their ranges or simply when they are measure in different measurement units </li>
    
<br>
    
<li> The Main Intention of Standardization is that it converts the :
    <ol>
        <b> <li>) Mean to 0 </b>
        <b> <li>) S.D. to 1 </b>
    </ol>
</li>
    
<br>

<li> Basically we are scaling down the values of our Dataset to a scale which will be pretty much similar to all the feature values, but <b> most impotantly keeping the RANGE SAME AS IN DATASET </b> </li>
    
</ul>

</p>

***

# Standardization Formula

<p>
    
Using this Formula we replace all the values individually we have in input with the Transformed/Recalculated values of the Z-Score
    
<ul>
    
<li> First we calculate the <b> Mean </b> of the Feature/Input Column </li>
         
    (As we already know it the Mean Value will always be a numerical Value middle/close to the middle of the values)
    

    
<li> Then we Calculate the <b> Standard Deviation </b> of the Feature/Input Column </li>

<br>
    
<li> <b> Finally we Substitute the Values of Mean and S.D. in the Standardization Formula </b> for each individuall value and calculate the Z-Score Individually for Each Value. </li>
    
</ul>

</p>

<p>

<b> NOTE :  </b>
    
<b> For Mean </b>

<ul>
<li> When we <b> Subtract a value Smaller than the Mean </b> we get <b> (-ve) Output </b> </li>

<li> When we <b> Subtract a value Larger than the Mean </b> we get <b> (+ve) Output </b> </li>  
</ul>

<br>

Hence, when we get (-ve) & (+ve) Values for <b> Subtraction of Value with Mean, </b> while Summation of all these values, <br>
<b> We get the Final Mean as 0. </b>

And when we get the Mean as 0, it means that <b> most or nearly all values are equal to to highly close to 0 and have very low variance among them. </b>

<b> Therefore, the S.D also becomes 1 (as good as no difference). </b>


</p>

***

# We'll check out the Basic Implementation

```python
dataset_0 = [10,5,6,1,3,7,9,4,8,2]
dataset_1 = [1,99,789,5,6859,541,94142,7,50826,35464]
n = len(dataset_1)
mean_ans = 0
ans = 0
j = 0

for i in dataset_1:
    j = j + i
    k = i*i;
    ans = ans + k   
    
print("n : ", n)
print("Summation (X)   : ", j)
print("Summation (X^2) : ", ans)
part_1 = ans/n
part_2 = mean_ans*mean_ans
standard_deviation = part_1 - part_2
print ("\nStandard Deviation : ", standard_deviation)

mean = j/n
print("Mean               : ", mean)
```

***

```python
final_z_score = []

print("\nZ-Score = (Current_Value - Mean) / S.D.\n")

for i in dataset_1:
    z_score = (i-mean)/standard_deviation
    final_z_score.append("{:.20f}".format(z_score))
    print("For", i)
    print("\t", i,"-",mean,"/", standard_deviation)
    print("\tZ - Score(", i, ") : {:.20f}".format(z_score))
    print("\n")
```

***

```python
print("  Original DataSet   |               Z-Score ")
print()
for i in range (len(dataset_1)):
    print("         ", dataset_1[i], "         |     ", final_z_score[i])
```

```python
#We are just using 10 values for our Dataset

#Here, dataset_0 will be constant as it's range is just 1 - 10
#But, dataset_1 will be scaled down as it's range is 1 - 95,000
global dataset_0, dataset_1
dataset_0 = [10,5,6,1,3,7,9,4,8,2]
dataset_1 = [1,99,789,5,6859,541,94142,7,50826,35464]

n = len(dataset_1)
mean_ans = 0
ans = 0
j = 0

for i in dataset_1:
    j = j + i
    k = i*i;
    ans = ans + k   
    
print('n : ', n)
print("Summation (X)   : ", j)
print("Summation (X^2) : ", ans)

part_1 = ans/n
part_2 = mean_ans*mean_ans
standard_deviation = part_1 - part_2
print ("\nStandard Deviation : ", standard_deviation)

mean = j/n
print("Mean               : ", mean)
print("\n\n\n")



#Calculating the Z-Score for each Value of dataset_1
final_z_score = []
print("Calculating Z-Score of Each Value in dataset_1")
for i in dataset_1:
    z_score = (i-mean)/standard_deviation
    final_z_score.append("{:.20f}".format(z_score))
    print(i,"-",mean,"/", standard_deviation)
    print("Z - Score(", i, ") : {:.20f}".format(z_score))
    print("\n")
    
    

#Comparing the Values of Original Dataset and Saled Down Dataset
#print("\n")
print("\nOriginal DataSet   |               Z-Score ")
print()
for i in range (len(dataset_1)):
    print("    ", dataset_1[i], "            |     ", final_z_score[i])
    
    
    
#Now we will compare and see the graph of the Original Values and the Standardized Values
import matplotlib 
import matplotlib.pyplot as plt

#Here We are checking the Graph of the Original Values
plt.scatter(dataset_0, dataset_1, label= "stars", color= "blue", marker= "*", s=40)
    
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('Original Values')
plt.legend()

plt.show()

#Here we are checking the Graph of the Standardized Values
plt.scatter(dataset_0, final_z_score, label= "stars", color= "blue", marker= "*", s=30)
  
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('Original Values')
plt.legend()
  
plt.show()
```
